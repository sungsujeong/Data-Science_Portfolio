## 08. Fitting Logistic Regression

```sas
/* Import dataset */
proc import 
    datafile='/home/u43674369/projects/project01/placement_data.csv' 
    out=df dbms=csv replace;
run;

/* Replace missing values with 0 */
proc stdize data=df
    out=df
    reponly missing=0;
run;
```

```sas
/* Fitting a binary Logistic regression model */
/* to characterize the relationship between */
/* variables, Status and Degree percentage */

proc format;
	value $job 'Placed' = 1 
			   'Not Placed' = 0 ;
run;

ods graphics on;

proc logistic data=df alpha=0.05
			  plots(only)=(effect oddsratio);
 	model Status(event='1') = Degree_P / clodds=pl;
 	format Status $job.; 
    title "LOGISTIC MODEL (1): Status = Degree_P";
run;
title;

ods graphics off;
```
