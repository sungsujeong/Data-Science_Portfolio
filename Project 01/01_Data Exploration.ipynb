# Project 01: Campust Recruitment Data Analysis
## 01. Data Exploration

```sas
/* Connect SAS */
sas_session = sas.SASsession()
sas_session
```

```sas
/* Import dataset */
proc import 
    datafile='/home/u43674369/projects/project01/placement_data.csv' 
    out=df dbms=csv replace;
run;
```

```sas
/* Check charicteristics of dataset and confirm the dataset correctly read */
proc contents data=df;
run;
```

```sas
proc print data=df;
    title "Campus Recruitment";
run;
```

```sas
/* Replace missing values with 0 */
proc stdize data=df
    out=df
    reponly missing=0;
run;

/* Check replaced missing values */
proc print data=df;
run;
```

```sas
/* Check frequencies of categorical variables in dataset */
proc freq data=df nlevels order=freq;
	table gender ssc_b hsc_b hsc_s workexp degree_t specialisation status;
	title "Frequencies of Categorical Variables";
run;
```
